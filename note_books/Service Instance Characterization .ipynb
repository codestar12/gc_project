### Way to many samples to plot easily. Need to sub sample

```python
service_instances = service_instances.sample(frac=.01, replace=False)
```

```python
from bokeh.plotting import figure, show
from bokeh.charts import Histogram
from bokeh.io import output_notebook
output_notebook()
```

## CPI might be a helpful measure of performance we need to maintain as if we make a model to prodict performance

```python
p = Histogram(service_instances["avg_cpi"])
show(p)
```

```python
p = Histogram(service_instances["cpu_util"])
show(p)
```

```python
p = Histogram(service_instances["mem_util"])
p.xaxis.axis_label = "% of requested memory used"
show(p)
```

```python
p = Histogram(service_instances["disk_util"])
p.xaxis.axis_label = "% of requested disk used"
show(p)
```
