```python
machine_service_df = pd.DataFrame()


for machine in machine_services[39600]:
    machine_service_df = machine_service_df.append(pd.DataFrame(machine_services[39600][machine]).transpose())

quantile_data = machine_service_df[np.logical_and(machine_service_df.load5 < machine_service_df.load5.quantile(.9), 
                                  machine_service_df.cpu_util < machine_service_df.cpu_util.quantile(.9))]

H, xedges, yedges = np.histogram2d(pd.to_numeric(quantile_data.load5/quantile_data.plan_cpu), pd.to_numeric(quantile_data.cpu_util), bins=10, range=[[0, .25],[0,19.103]])

H = H.T

plt.figure(figsize=[15,12])
plt.suptitle('Coustomer Facing Service Resource Utilization', fontsize=16)
plt.subplot(221)
plt.imshow(H, interpolation='nearest',origin='low', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]], aspect="auto")
plt.xlabel("linux cpu load")
plt.ylabel("cpu util %")
plt.title("11 AM")
plt.colorbar()

print(len(quantile_data.cpu_util))

#plt.show()
```
